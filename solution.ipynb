```python
import torch
import torch.nn as nn
```

```python
import yaml

# ÐšÐ¾Ð½Ñ„Ð¸Ð³ÑƒÑ€Ð°Ñ†Ð¸Ð¾Ð½Ð½Ñ‹Ð¹ Ñ„Ð°Ð¹Ð»
path = "./config.yaml"
with open(path, "r") as f:
    config = yaml.safe_load(f)
```

```python
from src.data_handle import process_data

train_loader, val_loader = process_data(config)
```

```python
from src.model import model_preparation

num_train_classes = len(train_loader.dataset.classes)
model = model_preparation(config, num_train_classes)
```

```python
from torch.optim import Adam
from src.one_epoch import TrainModel
from torch.optim.lr_scheduler import ReduceLROnPlateau

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model_preparation(config, num_train_classes).to(device)

criterion = nn.CrossEntropyLoss()

optimizer = Adam(model.parameters(), lr=config["train"]["lr"])
scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3)

train_model = TrainModel(config, model, optimizer, criterion, train_loader, val_loader, device, scheduler)

for epoch in range(1, config["train"]["epochs"] + 1):
    avg_train_loss = train_model.train(epoch)
    avg_vloss = train_model.val(epoch=epoch, avg_train_loss=avg_train_loss)
```

```python
train_model.eval()
```
